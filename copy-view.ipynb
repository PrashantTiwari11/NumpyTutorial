NumPy Array Copy vs View

```python
import numpy as np
```

```python
arr = np.array([1,23,34,55,66])
x = arr.copy()
```

```python
print(x)
```

```python
arr[0] = 42
```

```python
print(arr)
print(x)
```

```python
arr1 = np.array([1,2,3,4,5,6,7])
```

```python
y = arr1.view()
```

```python
arr1[2] = 34
```

```python
print(y)
```

```python
print(arr1)
```

```python
arr2 = np.array([1,2,3,4,5,6])
z = arr2.view()
```

```python
z[1] = 16
```

```python
print(z)
```

```python
print(arr2)
```

Using "base" we can check that is array owns the data or not.

```python
arr3 = np.array([1,2,3,4,5,5,6,7])
a = arr3.copy()
b = arr3.view()
```

```python
print(a.base)
```

```python
print(b.base)
```
